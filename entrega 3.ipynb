```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('rendimiento_2018.csv', delimiter=';')
data.head()
```

```python
data.columns
```

Se filtran las columnas que nos interesan para el problema planteado, las que en este caso son el establecimiento (RBD), si se encuentra en un lugar rural (RURAL_RBD), la asistencia del alumno, el género del alumno, la jornada del establecimiento (COD_JOR), el tipo de dependencia (COD_DEPE), el código de la comuna donde vive el estudiante, el código de la columna donde se encuentra el estblecimiento y el promedio general del estudiante. Luego de esto, se eliminaron aquellas filas que tenían promedio general igual a 0, ya que esto indica que la persona no realizó el año escolar.

```python
ic = data[[
    'RBD', 'RURAL_RBD', 'ASISTENCIA', 'GEN_ALU',
    'COD_JOR', 'COD_DEPE', 'COD_COM_ALU', 'COD_COM_RBD',
    'PROM_GRAL'
]]
ic = ic.drop(ic[ic['PROM_GRAL'] == "0"].index)
ic
```

```python
ic['PROM_GRAL'] = ic['PROM_GRAL'].apply(lambda x: x.replace(",", ".")).astype(np.float)
ic.astype(float)
```

Para seguir perfeccionando nuestro dataset, creamos una columna que muestra si el alumno vivía en la misma columna en la que se encuentra el establecimiento y separamos aquellas columnas que tenían diferentes valores de acuerdo a opciones (como el tipo de dependencia y jornada) en diferentes columnas binarias que mencionan si el estudiante cumple con la característica o no.

```python
regr_columns = ic[['RURAL_RBD', 'ASISTENCIA', 'GEN_ALU', 'PROM_GRAL']]
```

```python
regr_columns['SAME_COM'] = (ic.COD_COM_ALU == ic.COD_COM_RBD).astype(int)
depe = ['corp_munic', 'munic_daem', 'particular_sub', 'particular_pag', 'corp_admin_pag', 'serv_loc_ed']
for i in range(1, 7):
    regr_columns[depe[i - 1]] = (ic.COD_DEPE == i).astype(int)

jor = ['ma', 'tar', 'completa', 'vesp']
for i in range(1, 5):
    regr_columns[jor[i - 1]] = (ic.COD_JOR == i).astype(int)

regr_columns
```

Luego realizamos la regresión lineal con los datos que dejamos.

```python
from sklearn.linear_model import LinearRegression

regr = LinearRegression()
regr.fit(regr_columns.drop('PROM_GRAL', axis=1), regr_columns['PROM_GRAL'])

print(f'Coeficientes: {regr.coef_}')
print(f'Beta0: {regr.intercept_}')
```

```python
import json

regr_data = {
    'Coef': list(regr.coef_),
    'Beta': regr.intercept_
}

with open('E3/regr.json', 'w') as file:
    json.dump(regr_data, file)
```

```python
regr_columns.drop('PROM_GRAL', axis=1).columns[np.argmax(regr.coef_)]
```

A continuación mostraremos un ejemplo de 30 valores del dataset, para mostrar que funciona lo realizado.

```python
sample = regr_columns.sample(30)
sample
```

```python
regr.predict(sample.drop('PROM_GRAL', axis=1))
```

```python
prom_pred = regr.predict(regr_columns.drop('PROM_GRAL', axis=1))
```

Finalmente, vemos el error.

```python
from sklearn.metrics import mean_squared_error, r2_score

print(f'MSE: {mean_squared_error( regr_columns["PROM_GRAL"], prom_pred)}')
print(f'R2: {r2_score( regr_columns["PROM_GRAL"], prom_pred)}')
```
